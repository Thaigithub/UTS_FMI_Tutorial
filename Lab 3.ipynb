```python
def optimize_func(x:float, coupon: float, maturity: int, price: float):
  d = np.arange(start=1, stop=maturity+1)
  sum = np.sum(coupon/np.power(1+x,d))
  sum += 100/((1+x)**maturity)
  return price-sum
y_rate = np.array([])
for i in range(coupon.shape[0]):
  solver = root_scalar(optimize_func, args=(coupon[i], i+1, market_price[i]), bracket=[0.00001,1000])
  y_rate = np.append(y_rate, solver.root)

print(y_rate)
```

```python
df = pd.read_excel('AustralianGovernmentBonds.xlsx')
df[['coupon', 'maturity']] = df['Coupon and Maturity (click for term sheet)'].str.split(' ', n=1, expand=True)
df.drop(columns=['Coupon and Maturity (click for term sheet)'], inplace=True)
df.drop(columns=['ISIN'], inplace=True)
df.drop(columns=['Outstanding\n(face value, AUD million)'], inplace=True)
df.rename(columns={'Price': 'price'}, inplace=True)
df['coupon'] = df['coupon'].str.rstrip('%').astype(float) /100
df['maturity'] = pd.to_datetime(df['maturity'])
df
```

#### a.

```python
def coupon_payment(current_date: datetime.date, maturity_date: datetime.date, freq: int):
    checkdate = datetime.date(current_date.year+1 if maturity_date.year > current_date.year else current_date.year, maturity_date.month, maturity_date.day)
    n = (maturity_date.year - current_date.year)*freq
    while checkdate > current_date:
        n += 1
        checkdate = datetime.date(checkdate.year if checkdate.month>12/freq else checkdate.year-1, int(checkdate.month-12/freq) if checkdate.month>12/freq else 12 - int(12/freq - checkdate.month), checkdate.day)
    return (n if maturity_date.year == current_date.year else n-2 ,checkdate, datetime.date(checkdate.year if checkdate.month+12/freq < 12 else checkdate.year+1, int(checkdate.month+12/freq) if checkdate.month+12/freq < 12 else int(12/freq + checkdate.month) - 12, checkdate.day))


def optimize_func_config(y:float, coupon: float, price: float, n: int, w: float, freq: int, x: int):
  i = y/freq
  sum = (100/((1+i)**w)) * ((coupon/freq) * (x + (1-(1+i)**(1-n))/i) + 1/((1+i)**(n-1))) 
  return price-sum

y = np.array([])
date = datetime.date(2024, 3, 1)
freq = 2
for i in range(df['coupon'].size):
  n, last_payment, next_payment = coupon_payment(date, df['maturity'][i], freq)
  w = (next_payment-date).days/(next_payment-last_payment).days
  x = 1 if (next_payment-date).days > 7 else 0
  solver = root_scalar(optimize_func_config, args=(df['coupon'][i], df['price'][i], n, w, freq, x), x0=0.039)
  y = np.append(y, solver.root)
df['yield'] = y
df
```

#### b.

```python
if 'Zero Bond' in df.columns:
    del df['Zero Bond']

def coupon_payment_list(maturity: datetime.date, date: datetime.date, frequency: int):
    coupon_list = []
    checkdate = datetime.date(maturity.year, maturity.month, maturity.day)
    month = int(12/frequency)
    while (checkdate>date):
        coupon_list = [checkdate] + coupon_list
        checkdate = datetime.date( checkdate.year if checkdate.month > month else checkdate.year - 1, checkdate.month - month if checkdate.month > month else 12 + checkdate.month - month, checkdate.day)
    return coupon_list

def interpolation(today: datetime.date, last_date: datetime.date, last_df: float, next_date: datetime.date, next_df: float):
    return math.exp(math.log(last_df) + ((today-last_date).days/(next_date-last_date).days)*(math.log(next_df)-math.log(last_df)))

def interpolate_dates(date: datetime.date, date_array: list):
    greater_dates = date_array[date_array>date]
    next_bond = min(greater_dates, key=lambda x: abs(x - date)) if len(greater_dates)>0 else None
    lower_date = date_array[date_array<date]
    last_bond = min(lower_date, key=lambda x: abs(x - date)) if len(lower_date)>0 else None
    return [date, last_bond, next_bond]

def optimizer(price:float, coupon:float, interpolate: list, bond_price: pd.DataFrame, cb_price: float, today: datetime.date, maturity: datetime.date):
    if isinstance(price, np.ndarray):
        x = price[0]
    else:
        x = price
    factors = []
    for i in interpolate:
        if len(i) == 1:
            check = bond_price[bond_price['maturity']==i[0].strftime('%Y-%m-%d')]['Zero Bond']
            if len(check[check.isna()])==1:
                factors += [x/100]
            else:
                factors += [bond_price[bond_price['maturity']==i[0].strftime('%Y-%m-%d')]['Zero Bond'].iloc[0]/100]
        else:
            if i[2]==None:
                factors += [interpolation(i[0], i[1], bond_price[bond_price['maturity']==i[1].strftime('%Y-%m-%d')]['Zero Bond'].iloc[0]/100, maturity, x/100)]
            elif i[1]==None:
                factors += [interpolation(i[0], today, 1, i[2], bond_price[bond_price['maturity']==i[2].strftime('%Y-%m-%d')]['Zero Bond'].iloc[0]/100)]
            else:
                factors += [interpolation(i[0], i[1], bond_price[bond_price['maturity']==i[1].strftime('%Y-%m-%d')]['Zero Bond'].iloc[0]/100, i[2], bond_price[bond_price['maturity']==i[2].strftime('%Y-%m-%d')]['Zero Bond'].iloc[0]/100)]
    factors = np.array(factors)
    return np.sum(factors*100*coupon) + x - cb_price


df['Zero Bond'] = np.nan
today = datetime.date(2024,3,1)
for i in range(len(df)):
    bond_maturities = df[df['Zero Bond'].isna()==False]['maturity'].dt.date.array
    coupon_dates = coupon_payment_list(df['maturity'].dt.date[i], today, 2)
    interpolate = []
    for j in coupon_dates:
        if j in bond_maturities or len(df[df['maturity']==j.strftime('%Y-%m-%d')])!=0:
            interpolate += [[j]]
        else:
            interpolate += [interpolate_dates(j, bond_maturities)]
    solver = root_scalar(optimizer, args=(df['coupon'][i]/2, interpolate, df[['Zero Bond', 'maturity']], df['price'][i], today, df['maturity'].dt.date[i]), x0=100, method='newton')
    df.at[i, 'Zero Bond'] = solver.root
df.to_excel('Tutorial3_Answer.xlsx', index=False)
df
```

```python
def Yield(y,dtn,dpn,c,n,P):
    if dtn <= 7:
        k=0
        
    else:
        k=1
    return (1+y/2)**(-dtn/dpn) * (c*k + c*(1-(1+y/2)**(1-n))/(y/2) + 100*(1+y/2)**(1-n)) - P
```
